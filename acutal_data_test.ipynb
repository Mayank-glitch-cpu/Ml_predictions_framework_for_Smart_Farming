```python
from sklearn.linear_model import Ridge
rr= Ridge(alpha=0.1)
```

```python
predictors= wheat_data.columns[~wheat_data.columns.isin(['target'])]
```

```python
predictors
```

```python
train= wheat_data.iloc[:1369,:]
test= wheat_data.iloc[1369:1379,:]
rr.fit(train[predictors], train["target"])
        
preds = rr.predict(test[predictors])
preds = pd.Series(preds, index=test.index)
print(train)
print(test['T'])
print(preds)
```

```python
def backtest(data, model, predictors, start=1369, step=10):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[:i,:]
        test = data.iloc[i:(i+step),:]
        
        model.fit(train[predictors], train["target"])
        
        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual_temp", "predicted_temperature"]
        combined["diff"] = (combined["predicted_temperature"] - combined["actual_temp"]).abs()
        
        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions= backtest(wheat_data,rr,predictors)
```

```python
predictions=predictions.ffill()
predictions
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(predictions['actual_temp'],predictions['predicted_temperature'])
```
